```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value, plot_confusion_matrix
from jcopml.feature_importance import mean_score_decrease

import matplotlib.pyplot as plt
import seaborn as sns

pd.options.display.max_columns = 999
```

# Import Datasets

```python
df = pd.read_csv("datasets/weatherAUS.csv", parse_dates=["Date"])
```

Lakukan Ekstraksi pada kolom Date

```python
# buat kolom year yang berisi data tahun hasil ekstraksi kolom date
df['year'] = pd.DatetimeIndex(df['Date']).year

# buat kolom month yang berisi hasil ekstraksi bulan pada kolom date
df['month'] = pd.DatetimeIndex(df['Date']).month

# buat kolom day yang berisi hari
df['day'] = pd.DatetimeIndex(df['Date']).day

# hapus kolom Date karena sudah tidak terpakai
df = df.drop(columns="Date")

# tampilkan dataframe
df.head()
```

# Datasets Splitting

```python
X = df.drop(columns="RainTomorrow")
y = df["RainTomorrow"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Simple Model
### Preprocessor

```python
from xgboost import XGBClassifier
```

```python
preprocessor = ColumnTransformer([
    ('numeric', num_pipe(scaling='standard',transform='yeo-johnson'), X_train.describe().columns),
    ('categoric', cat_pipe(encoder='onehot'), X_train.select_dtypes('object').columns)
])

pipeline = Pipeline([
    ('prep', preprocessor),
    ('algo', XGBClassifier(n_jobs=-1, random_state=42))
])
```

### Training

```python
from sklearn.model_selection import RandomizedSearchCV
from jcopml.tuning import random_search_params as rsp
from jcopml.tuning.space import Integer, Real
```

```python
model = RandomizedSearchCV(pipeline, rsp.xgb_params, cv=5, n_iter=5, n_jobs=-1, verbose=1, random_state=42)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```

### Scoring

```python
from sklearn.metrics import f1_score, recall_score
```

```python
plot_confusion_matrix(X_train, y_train, X_test, y_test, model)
```

```python
recall_score(y_test, model.predict(X_test), pos_label='No')
```

```python
f1_score(y_test, model.predict(X_test), pos_label='Yes')
```

# Features Importance

```python
mean_score_decrease(X_train, y_train, model, plot=True, topk=10)
```

# Tuning - 1

```python
parameter_tune = {
    'algo__max_depth': Integer(low=1, high=10),
    'algo__learning_rate': Real(low=-2, high=0, prior='log-uniform'),
    'algo__n_estimators': Integer(low=100, high=200),
    'algo__subsample': Real(low=0.3, high=0.8, prior='uniform'),
    'algo__gamma': Integer(low=1, high=10),
    'algo__colsample_bytree': Real(low=0.1, high=1, prior='uniform'),
    'algo__reg_alpha': Real(low=-1, high=2, prior='log-uniform'),
    'algo__reg_lambda': Real(low=0, high=2, prior='log-uniform')}
```

```python
model = RandomizedSearchCV(pipeline, parameter_tune, cv=5, n_iter=5, n_jobs=-1, verbose=1, random_state=42)
model.fit(X_train, y_train)

print(model.best_params_)
print(model.score(X_train, y_train), model.best_score_, model.score(X_test, y_test))
```
